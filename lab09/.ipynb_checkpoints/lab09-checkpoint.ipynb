## Zadanie 1. Prosty ranking wierzchołków.

```python
def normalize_matrix(A):
    N = len(A)
    V = np.zeros(N)
    
    for u in range(N):
        for v in range(N):
            if A[u][v] != 0:
                V[v] += 1
                
    A /= V
    
    return A

def pagerank_1(A, max_iterations=32, d=0.8):
    A = normalize_matrix(A)
    N = A.shape[1]
    r = np.ones(N) / N
    
    for _ in range(max_iterations):
        r = d * A @ r
        if (np.absolute(r) < 10*-7).all():
            break
        
    return v

A = np.random.rand(256, 256)
pagerank_1(A)
```

## Zadanie 2.

```python
def pagerank_2(A, max_iterations=32, d=0.8):
    A = normalize_matrix(A)
    N = A.shape[1]
    v = np.ones(N) / N
    r = (d * A + (1 - d) / N)
    
    for _ in range(max_iterations):
        v = r @ v
        if (np.absolute(v) < 10*-7).all():
            break
        
    return v

A = np.random.rand(256, 256)
pagerank_2(A)
```
