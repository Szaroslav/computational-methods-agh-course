## Zadanie 1. Przekształcenie sfery w elipsoidę.

1. Korzystając z równania parametrycznego narysuj sferę jednostkową w 3D
$$
v = \begin{bmatrix}
    \cos{(s)}\sin{(t)} \\
    \sin{(s)}\sin{(t)} \\
    \cos{(t)}
\end{bmatrix} \\
s \in [0, 2\pi],\; t \in [0, \pi]
$$

```python
def get_sphere_points(n):
    s_space = np.linspace(0, 2 * np.pi, n)
    t_space = np.linspace(0, np.pi, n)
    
    return np.array([(np.cos(s) * np.sin(t), np.sin(s) * np.sin(t), np.cos(t)) for s in s_space for t in t_space])

def plot_scatter(x, y, z):
    fig = plt.figure()
    ax = fig.add_subplot(projection='3d')

    ax.scatter(x, y, z)
    ax.set_box_aspect((1, 1, 1))

    plt.show()
```

```python
points = get_sphere_points(2**4)
plot_scatter(points[:, 0], points[:, 1], points[:, 2])
```
